# **<font color="77a353"> Obtendo os dados**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from IPython.display import display, Markdown
```

```python
dados = pd.read_json('https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json')
dados.head()
```

```python
dados.iloc[0]
```

```python
# Expandindo cada coluna aninhada
customer_df = pd.json_normalize(dados['customer'])
phone_df    = pd.json_normalize(dados['phone'])
internet_df = pd.json_normalize(dados['internet'])
account_df  = pd.json_normalize(dados['account'])

# Adicionando prefixos para evitar nomes repetidos
customer_df.columns = [f'customer_{col}' for col in customer_df.columns]
phone_df.columns    = [f'phone_{col}' for col in phone_df.columns]
internet_df.columns = [f'internet_{col}' for col in internet_df.columns]
account_df.columns  = [f'account_{col}' for col in account_df.columns]

# Juntando tudo com as colunas principais
dados_limpo = pd.concat(
    [
        dados[['customerID', 'Churn']].reset_index(drop=True),
        customer_df,
        phone_df,
        internet_df,
        account_df
    ],
    axis=1
)
```

# <font color="77a353">**Padronização e Transformação de Dados**

```python
dados_limpo.info()
```

```python
# Lista de colunas que você quer transformar em binário
colunas_binarias = [
    'Churn',
    'customer_Partner',
    'customer_Dependents',
    'phone_PhoneService',
    'phone_MultipleLines',
    'internet_OnlineBackup',
    'internet_DeviceProtection',
    'internet_TechSupport',
    'internet_StreamingTV',
    'internet_StreamingMovies',
    'account_PaperlessBilling'
]

# Passo 1: padronizar os valores para minúsculas e remover espaços
for col in colunas_binarias:
    dados_limpo[col] = dados_limpo[col].str.strip().str.lower()

# Passo 2: converter 'yes' para 1 e 'no' para 0
# Se houver outros valores como 'no internet service', podemos converter também
mapeamento = {
    'yes': 1,
    'no': 0
}

# Aplicar o mapeamento
for col in colunas_binarias:
    dados_limpo[col] = dados_limpo[col].map(mapeamento)
```

A análise explora o comportamento de clientes que **permaneceram** versus os que **cancelaram** o serviço, gerando

- **Qualidade**: checagem de **nulos** e **duplicatas**; remoções/ajustes quando necessário.  
- **Criação de feature**: `Contas_Diarias = (mensal / 30)` para granularidade diária.

##  Conclusões e Insights
**Padrões observados (exemplos típicos nesse tipo de base):**
- **Contrato**: planos **mensais** tendem a apresentar **maior proporção** de evasão em comparação a contratos mais longos (12/24 meses).  
- **Pagamentos**: métodos com **maior atrito** (ex.: boleto/manual) costumam ter evasão maior do que **débito automático/crédito**.  
- **Tempo de casa (tenure)**: clientes **recém-adquiridos** (tenure baixo) usualmente concentram mais churn; retenção melhora conforme o relacionamento amadurece.  
- **Gasto mensal/total**: variação pode indicar **sensibilidade a preço** ou impacto de **addons/serviços** (streaming, segurança, etc.).



##  Recomendações  
1. **Ofertas de fidelização** para clientes em **contratos mensais**: upgrade para planos anuais com **desconto/espaçamento de parcelas**.  
2. **Incentivar métodos automáticos** (débito/CRÉDITO): bônus na primeira fatura e comunicação ativa para migração.  
3. **Onboarding reforçado** nos **primeiros meses**: check-ins proativos, tutoriais e suporte priorizado para reduzir churn inicial.  
4. **Alertas de risco**: monitore sinais como **aumento de chamados**, **inadimplência** e **redução de uso** para acionar time de retenção.  
5. **Testes A/B**: experimente **pacotes de valor** (combo internet + segurança + streaming) e **ofertas segmentadas** por perfil (ex.: heavy streamers).
